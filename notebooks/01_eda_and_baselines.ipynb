The Preprocessing Steps:

To make the text digestible for our classical NLP algorithms, we need to do three things in our notebook:

Lowercase everything: So "The" and "the" are treated as the same word.

Remove punctuation: We want to strip out commas, periods, and question marks (like the ? in row 0) so we just focus on the words.

Tokenize: Split the continuous string of text into a list of individual words.

Binarize the Labels: Machine learning models prefer numbers, so we will map 'student' to 0 and 'ai' to 1, just like we outlined for your Day 7 feature matrix.

```python
import re
import nltk
from nltk.tokenize import word_tokenize
```

```python
# nltk.download('punkt')
# nltk.download('punkt_tab')
```

```python
# 1. Assign your features and map labels to 0 and 1
X = df['Text']
y = df['Label'].map({'student': 0, 'ai': 1})
```

```python
# 2. Create the cleaning and tokenization function
def clean_and_tokenize(text):
    # Convert to string and lowercase
    text = str(text).lower()
    
    # Remove punctuation and numbers (keep only lowercase letters and spaces)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Tokenize into words
    tokens = word_tokenize(text)
    
    return tokens
```

```python
# 3. Apply the function to the entire Text column
print("Tokenizing data...")
X_processed = X.apply(clean_and_tokenize)
```
